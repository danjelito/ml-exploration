```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.svm import SVC

from plotting import plot_decision_boundary
```

```python
x, y = make_moons(n_samples=200, noise=0.2, random_state=42)
x.shape, y.shape
```

```python
plt.scatter(x[:, 0], x[:, 1], c=y)
plt.plot()
```

# Different kernels

```python
kernels = ["linear", "poly", "rbf", "sigmoid"]

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))

for kernel, ax in zip(kernels, axes.ravel()):
    model = SVC(kernel=kernel)
    model.fit(x, y)
    plot_decision_boundary(model, x, y, ax=ax, plot_title=f"{kernel} kernel")

plt.tight_layout()
plt.show()
```

# Value of C

```python
start = 0.00001
step_count = 8
cs = start * (10 ** np.arange(step_count))

fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(12, 6))

for c, ax in zip(cs, axes.ravel()):
    model = SVC(kernel="linear", C=c)
    model.fit(x, y)
    plot_decision_boundary(model, x, y, ax=ax, plot_title=f"kernel linear, c {c: .5f}")

plt.tight_layout()
plt.show()
```

# Value of degree for poly kernel

```python
degrees = np.arange(1 , 11, 1)

fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(12, 6))

for degree, ax in zip(degrees, axes.ravel()):
    model = SVC(kernel="poly", degree=degree)
    model.fit(x, y)
    plot_decision_boundary(model, x, y, ax=ax, plot_title=f"kernel poly, degree {degree: .0f}")

plt.tight_layout()
plt.show()
```
